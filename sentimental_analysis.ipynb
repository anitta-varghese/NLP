```python
from textblob import TextBlob
import pandas as pd

reviews = [
    "This phone is absolutely amazing! Best purchase ever!",
    "Battery life is terrible. Dies in 2 hours.",
    "Good phone but a bit expensive for what it offers.",
    "Camera quality is outstanding! Love taking photos with it.",
    "Worst phone I've ever owned. Constantly crashes.",
    "It's okay. Nothing special but does the job.",
    "Incredible value for money! Highly recommend!",
    "Screen broke after one week. Very disappointed.",
    "Fast performance and beautiful design. Very happy!",
    "Customer service was unhelpful when I had issues."
]
sentiments = []
polarities = []
categories = []
for review in reviews:
  blob = TextBlob(review)
  polarity = blob.sentiment.polarity
  print(blob,blob.sentiment.polarity)

  if polarity > 0.1:
        category = "Positive"
  elif polarity < -0.1:
        category = "Negative"
  else:
        category = "Neutral"

sentiments.append(review)
polarities.append(polarity)
categories.append(category)

```
