```python
#Process 2
import pandas as pd

data = {
    "text": [
        "i hate monotonous life", "but i love living", "but its confusing"
    ]
}

df = pd.DataFrame(data)
df

import nltk
nltk.download('vader_lexicon')

from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

df['score'] = df['text'].apply(lambda x: sia.polarity_scores(x)['compound'])
df['sentiment'] = df['score'].apply(lambda x:
                                    'Positive' if x > 0
                                    else ('Negative' if x < 0
                                    else 'Neutral'))
df
```

how VADER differs form Affin and TextBlob?  VADER, AFINN, and TextBlob all detect the first sentence as negative and the second as positive, but differ slightly on the third.
AFINN gives simple word-based scores, while TextBlob adds subjectivity but ignores connectors like “but.”
VADER performs best overall because it accounts for negation, intensity, and contrast words, capturing the mixed feeling in “but it’s confusing.”

```python

import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from afinn import Afinn
from textblob import TextBlob

# Download VADER lexicon if not already downloaded
try:
    nltk.data.find('sentiment/vader_lexicon.zip')
except nltk.downloader.DownloadError:
    nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()
af = Afinn()

sentences = ["Wow, this meeting was just *so productive*... I totally didn’t waste two hours 😒"]

# Create DataFrame
df = pd.DataFrame(sentences, columns=['text'])

# Apply AFINN, TextBlob, and VADER sentiment analysis
df['afinn_score'] = df['text'].apply(lambda x: af.score(x))

df['textblob_polarity'] = df['text'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['textblob_subjectivity'] = df['text'].apply(lambda x: TextBlob(x).sentiment.subjectivity)
df['textblob_sentiment'] = df['textblob_polarity'].apply(
    lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral')
)

df['vader_scores'] = df['text'].apply(lambda x: sia.polarity_scores(x))
df['vader_compound'] = df['vader_scores'].apply(lambda x: x['compound'])
df['vader_sentiment'] = df['vader_compound'].apply(lambda x:
                                    'Positive' if x > 0
                                    else ('Negative' if x < 0
                                    else 'Neutral'))

# Display the results
display(df[['text', 'afinn_score', 'textblob_polarity', 'textblob_subjectivity', 'textblob_sentiment', 'vader_compound', 'vader_sentiment']])
```

```python
### TASK 2

texts=[
    "I LOVE this movie!!! 😍😍",
    "This course is SO boring... 😴 ",
    "I think AI is amazing 🤖💡",
    "Not sure if I like this. 🤷",
    "Great work!!! Keep it up 🔥🔥"
]
import pandas as pd
df= pd.DataFrame(texts, columns=["texts"])
df

df['clean_text'] = df['texts'].str.lower()
df['clean_text'] = df['clean_text'].str.strip()
df[['texts','clean_text']]

import string
df['no_punct'] = df['clean_text'].str.translate(str.maketrans('', '', string.punctuation))
df[['clean_text','no_punct']]

nltk.download('stopwords')
from nltk.corpus import stopwords

stop = set(stopwords.words('english'))
df['no_stopwords'] = df['no_punct'].apply(lambda x: " ".join([word for word in x.split() if word not in stop]))
df[['no_punct','no_stopwords']]

import re

def remove_emojis(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"
                           u"\U0001F300-\U0001F5FF"
                           u"\U0001F680-\U0001F6FF"
                           u"\U0001F1E0-\U0001F1FF"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

df['no_emoji'] = df['no_stopwords'].apply(remove_emojis)
df[['no_stopwords','no_emoji']]

```
